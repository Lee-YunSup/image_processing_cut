## x좌표, y좌표 append 문제 (setMouseCallback)

```python

list1=[]
x_point=[]
y_point=[]

def paint_draw(event,former_x,former_y,ddd,ddda):

    if event==cv2.EVENT_LBUTTONDOWN: # 왼쪽 마우스 클릭
        x_point.append(former_x)
        y_point.append(former_y)

    elif event==cv2.EVENT_MOUSEMOVE: # 마우스 움직임
                x_point.append(former_x)
                y_point.append(former_y)
                
    elif event==cv2.EVENT_LBUTTONUP: # 왼쪽 마우스 때기            
            x_point.append(former_x)
            y_point.append(former_y)
    list1.append([former_x,former_y])
    cv2.imshow('image', image)
```

처음 코드를 작성할 때 좌표값을 구하기 위해 list1.append를 각 조건문 밖에 작성을 해주니, 위 마우스 클릭이나 마우스 움직임을 하지않고, 이미지 위에 올려만 놓아도 cv2.setMouseCallback 함수를 사용했을 때, 좌표 값이 list로 추가 되었다.

그래서 list1.append함수를  각각 조건문 안에 사용해주니 의도에 맞게 동작하였다.
